```python
import pandas as pd
from datetime import datetime as dt
```

```python
si = pd.read_pickle('data/en/outputs/thank_df.pickle')
di = si.set_index('timestamp')
di.sort_index(inplace=True)
```

```python
di.head()
```

```python
di['timestamp']=di.index
```

```python
di.head()
```

```python
di.index.min()
```

```python
di[di.index.max():].shape[0] -1
```

```python
si.shape
```

```python
dt1 = dt(2016, 6, 1)
dt2 = dt(2016, 7, 1)
```

```python
%%timeit -n3
cond1 = si['timestamp'] > dt1
cond2 = si['timestamp'] <= dt2
res = si[(cond1) & (cond2)]
print(res.shape[0])
```

```python
%%timeit -n3
res = di[dt1:dt2]
print(res.shape[0])
```
